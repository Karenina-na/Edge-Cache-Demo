# 图30 分类图

分类图可用于可视化彼此相关的2个或更多分类变量的计数分布，seaborn直接支持分类图.

```python
from doc.Plot.require import *

# Load Dataset
titanic = sns.load_dataset("titanic")

g = sns.catplot(x = "alive", col="deck", col_wrap=4, data=titanic[titanic.deck.notnull()], kind="count", height=3.5, aspect=0.8, palette='tab20')

plt.show()
```
